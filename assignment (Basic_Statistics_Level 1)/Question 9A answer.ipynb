```python
import pandas as pd
import numpy as np
import math
from scipy.stats import skew, kurtosis
```

```python
df = pd.read_csv('Q9_a.csv')
```

```python
df.head()
```

## Skewness

```python
speed_skewness = skew(df['speed'])
```

```python
speed_skewness
```

```python
distance_skewness = skew(df['dist'])
```

```python
distance_skewness
```

# kurtosis

```python
speed_kurtosis = kurtosis(df['speed'])
```

```python
speed_kurtosis
```

```python
distance_kurtosis = kurtosis(df['dist'])
```

```python
distance_kurtosis
```
