```python
import pandas as pd
import numpy as np
import math
from scipy.stats import skew, kurtosis
```

```python
df = pd.read_csv('Q9_b.csv')
```

```python
df.head()
```

## Skewness

```python
speed_skewness = skew(df['SP'])
```

```python
speed_skewness
```

```python
distance_skewness = skew(df['WT'])
```

```python
distance_skewness
```

# kurtosis

```python
speed_kurtosis = kurtosis(df['SP'])
```

```python
speed_kurtosis
```

```python
distance_kurtosis = kurtosis(df['WT'])
```

```python
distance_kurtosis
```
